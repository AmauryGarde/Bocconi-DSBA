```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver 
import time
import csv
```

```python
#csv file setup
with open("output-file.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Designer", "Product", "Description"])
    
#driver setup
driver = webdriver.Chrome(executable_path = r"C:\Users\garde\Desktop\chromedriver.exe")
url = "https://www.myminifactory.com/category/accessibility"
driver.get(url)
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

#scrolling block
last_height = driver.execute_script("return document.body.scrollHeight")
SCROLL_PAUSE_TIME = 0.5

while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(SCROLL_PAUSE_TIME)
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
    
#scraping setup
content = driver.page_source
soup = BeautifulSoup(content, "lxml")
information = soup.find_all("div", class_="search-result-each")

#scraping href
URLs = []
for product in information:
    page_url = product.find("a").get("href")
    page_url = "https://www.myminifactory.com" + str(page_url)
    URLs.append(page_url)
    
#scraping relevant data
for url in URLs:
    driver.get(url)
    content = driver.page_source
    soup = BeautifulSoup(content, "lxml")
    
    designers = soup.find_all("div", class_="row rowUserInfo")
    for designer in designers:
        designer_info = designer.get_text()
        designer_name = designer_info.splitlines()[3]
        
    products = soup.find_all("h1", class_="obj-title show-for-large-up")
    for product in products:
        product_info = product.get_text()
        product_name = product_info.splitlines()[1].replace('\t', '')
        
    descriptions = soup.find_all("div", class_="short-text text-auto-link")
    for description in descriptions:
        description_info = description.get_text()
    
    #storing data
    with open("output-file.csv", "a", newline="", encoding="utf-32") as f:
            writer = csv.writer(f)
            writer.writerow([designer_name, product_name, description_info])

driver.close()
```
